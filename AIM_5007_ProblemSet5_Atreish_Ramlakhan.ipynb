```python
data = pd.read_csv('https://gist.githubusercontent.com/keeganhines/59974f1ebef97bbaa44fb19143f90bad/raw/d9bcf657f97201394a59fffd801c44347eb7e28d/Hitters.csv',index_col=[0])
data.head(5)
```

```python
data.shape
```

```python
data.reset_index(inplace = True)
```

```python
data1 = data.drop(columns=['index'])
```

```python
#apply one hot encoding for object datatype columns
encoded_data = pd.get_dummies(data1)
```

```python
data2 = encoded_data.dropna(how='any')
data2.shape
```

```python
X = data2.drop(columns=['Salary'])
y = data2.Salary
```

```python
seed = 1
np.random.seed(seed)
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print ("Number of examples in training set = ", X_train.shape[0])
print ("Number of examples in test set = ", X_test.shape[0])
```

```python
# define the keras model
model = Sequential()
model.add(Dense(units = 10, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units = 8, activation='relu'))
model.add(Dense(units = 8, activation='relu'))
model.add(Dense(units = 8, activation='relu'))
model.add(Dense(units = 8, activation='relu'))
model.add(Dense(units = 8, activation='relu'))
model.add(Dense(units = 1, activation='relu'))
model.compile(optimizer=Adam(learning_rate=0.1), loss='mse', metrics = ['mse'])
model.summary()
```

```python
#model.fit(X, y, batch_size=5, epochs=200, verbose=1, validation_split=0.2, shuffle=True)

model.fit(X, y, epochs=200, verbose=1, shuffle=False)
```

```python
seeds = pd.read_csv('https://raw.githubusercontent.com/atreish/AIM_5007_Neural_Networks_-_Deep_Learning/main/seeds_dataset.txt',
                   sep='\s+',header=None, names = ['Area','Perimeter','Compactness','Klength','Kwidth','Asym_coeff','length of kernel groove','class_label'])
seeds.head(5)
```

```python
#Normalize Numeric Data, Not the label column
seeds.iloc[:,:-1] = seeds.iloc[:,:-1].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

#Shuffle the dataset to ensure the network is trained on all 3 available classes
seeds1 = seeds.sample(frac=1).reset_index(drop=True)

#Change X  (numeric columns) and y (class columns) to arrays and the target datatype to int
X = np.array(seeds1)[:, :-1]
y = np.array(seeds1)[:, -1].astype(int)

#Apply one hot encoding for each class, this will be our target class 
target_layer = []
for i in y:
    if i == 1:
        target_layer.append([1,0,0])
    elif i == 2:
        target_layer.append([0,1,0])
    else:
        i == 3
        target_layer.append([0,0,1])
        
target_layer = np.array(target_layer)
```

```python
model1 = Sequential()
model1.add(Dense(units = 7, activation='relu', input_dim=X_train.shape[1]))
model1.add(Dense(units = 7, activation='relu'))
model1.add(Dense(units = 3, activation='softmax'))
model1.compile(optimizer=Adam(learning_rate=.1), loss='categorical_crossentropy', metrics=['accuracy'])
model1.summary()
```

```python
model1.fit(X, target_layer, epochs=200, verbose=1, shuffle=False)
```
