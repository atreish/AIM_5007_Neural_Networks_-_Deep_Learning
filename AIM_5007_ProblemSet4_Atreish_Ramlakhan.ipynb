# **1. HYPERBOLIC TANGENT**

$ f(x) = tanh(x) = \dfrac{e^x - e^{-x}}{e^x + e^{-x}} $

**I. Sketch the function**

````python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Hyperbolic Tangent Activation Function')
plt.show()
````

**II. Differentiate the function**

$ \dfrac{dy}{dx} tanh(x)' $

$ = \dfrac{sinh(x)'}{cosh(x)'} $

$ = \dfrac{sinh(x)'cosh(x)-sinh(x)cosh(x)'}{cosh^2x} $

$ = \dfrac{cosh(x)cosh(x)-sinh(x)sinh(x)}{cosh^2x} $

$ = \dfrac{cosh^2(x)-sinh^2(x)}{cosh^2(x)} $

$ = \dfrac{1}{cosh^2(x)} = sech(x) $

**III. Identify any tricky points in taking the derivative (i.e., discontinuities) and propose a reasonable solution to the difficulty (i.e., assign a value for that point)**

```The Hyperbolic Tangent function is a smooth and continuous function. It is the quotient of 2 continuous functions sinh(x) and cosh(x). With that, there will be no issue with taking dervitives from (-inf, inf)```

**IV. Sketch the derivitive**

````python
def sech(x):
    return 1 - (np.e**x-np.e**-x)**2/(np.e**x+np.e**-x)**2 

x = np.linspace(-10, 10)
plt.plot(x, sech(x))
plt.axis('tight')
plt.title('Derivitive of Hyperbolic Tangent ')
plt.show()
````

# **2. LINEAR FUNCTION**

$ f(x) = kx $

**I. Sketch the function**

````python
def linearfun(k,x):
    return k*x

x = np.linspace(-10, 10)
k_list = [-2,-1,0,1,2]
for k in k_list:
    plt.plot(x, linearfun(k,x))
    plt.axis('tight')
    plt.title('Linear Activation Function. k={}'.format(k))
    plt.show()
````

**II. Differentiate the function**

```The derivitive is simply the input k```

$ \dfrac{dy}{dx} [k*x] = k $

**III. Identify any tricky points in taking the derivative (i.e., discontinuities) and propose a reasonable solution to the difficulty (i.e., assign a value for that point)**

```The Linear function is continuous. There are no loops or discontinuities. If the k is 0, there we have a line at the x axis, but if there is a constant term c as in y=mx + c, then the line will be at y=c```

**IV. Sketch the derivitive**

````python
plt.axhline(y=3, color='r', linestyle='-')
plt.xlim((-10, 10))
plt.axis('tight')
plt.title('Derivitive of Linear Funation')
plt.show()
````

# **3. STEP FUNCTION**

$
f(x) = \left\{
    \begin{array}\\
        1 & \mbox{if } \ 𝑥 ≤ 0 \\
        0 & \mbox{if } \ 𝑥 > 0 \\
    \end{array}
\right.
$

**I. Sketch the function**

````python
plt.figure(figsize=(5, 5))
y = np.array([0,0,1,1])
x = np.array([-10,0,1,10])
plt.step(x, y)
plt.xlim((-3, 3))
plt.ylim((-3, 3))
plt.show()
````

**II. Differentiate the function**

```The derivitive is 0 at both ends of the function. However there exists a discontinuity at x=0```

**III. Identify any tricky points in taking the derivative (i.e., discontinuities) and propose a reasonable solution to the difficulty (i.e., assign a value for that point)**

```Since the step function is mostly a continuous line, i.e. a constant, the derivitive will be 0 for both "sections" of the data. However, there is a discontinuity at x=0. The "jump" in the graph, in mathematical terms is not defined, and in fact does not exist. I would replace this is with 0 to make it consistent for all real numbers```

**IV. Sketch the derivitive**

````python
plt.axhline(y=0, color='r', linestyle='-')
plt.xlim((-10, 10))
plt.axis('tight')
plt.title('Derivitive of Step Function')
plt.show()
````

# **4. ReLU Function**

$ ReLU(x) = max(0,x) $

**I. Sketch the function**

````python
def relu(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1

x = np.linspace(-10, 10)
plt.plot(x, relu(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
````

**II. Differentiate the function**

```The derivitive is 0 where x<0 and it is 1 where x>0```

$
ReLU'(x) = \left\{
    \begin{array}\\
        0 & \mbox{if } \ 𝑥 < 0 \\
        1 & \mbox{if } \ 𝑥 > 0 \\
    \end{array}
\right.
$

**III. Identify any tricky points in taking the derivative (i.e., discontinuities) and propose a reasonable solution to the difficulty (i.e., assign a value for that point)**

```Since the right and left portion of the function does not match, we have another discontinuity at x=0. This is a trickier type of problem as both left and right limits approach an undefined value but for simplicity sake I would define it as ReLU'(x)=0 just to retain continuity```

**IV. Sketch the derivitive**

````python
def drelu(x):
    x2=[]
    for i in x:
        if i<=0: #this is decided so that we can retain continuity
            x2.append(0)
        else:
            x2.append(1)
    return x2

x = np.linspace(-10, 10)
plt.plot(x, drelu(x))
plt.axis('tight')
plt.title('Derivitive of RuLU Function')
plt.show()
````

# **5. SoftMax Function**

$ 𝑓(𝑥) = ln (1 + 𝑒^𝑥) $

**I. Sketch the function**

````python
def softplus(x):
    return np.log(1 + np.e**x)

x = np.linspace(-10, 10)
plt.plot(x, softplus(x))
plt.axis('tight')
plt.title('Activation Function: Softmax')
plt.show()
````

**II. Differentiate the function**

$  \dfrac{dy}{dx} ln(1+e^x) $

$ = \dfrac{1}{e^x + 1} \dfrac{dy}{dx} e^x + 1 $

$ = \dfrac{\dfrac{d}{dx} e^x + \dfrac{d}{dx} 1}{e^x + 1} $

$ = \dfrac{\mathrm{e}^x}{\mathrm{e}^x+1} $


**III. Identify any tricky points in taking the derivative (i.e., discontinuities) and propose a reasonable solution to the difficulty (i.e., assign a value for that point)**

```This is a continuously smooth function over all real numbers. there are no discontinuities```

**III. Sketch the derivitive**

````python
def dsoftplus(x):
    return np.e**x / (1+np.e**x)

x = np.linspace(-10, 10)
plt.plot(x, dsoftplus(x))
plt.axis('tight')
plt.title('Derivitive of SoftPlus Function')
plt.show()
````
